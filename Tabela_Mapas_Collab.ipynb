```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Lista de locais
locais = [
    {"nome_local": "Palco Pop", "latitude": -23.5820, "longitude": -46.6340},
    {"nome_local": "Palco Eletrônico", "latitude": -23.5825, "longitude": -46.6330},
    {"nome_local": "Palco Rock", "latitude": -23.5815, "longitude": -46.6325},
    {"nome_local": "Palco Secundário A", "latitude": -23.5830, "longitude": -46.6350},
    {"nome_local": "Palco Secundário B", "latitude": -23.5840, "longitude": -46.6345},
    {"nome_local": "Área de convivência 1", "latitude": -23.5845, "longitude": -46.6335},
    {"nome_local": "Área de convivência 2", "latitude": -23.5850, "longitude": -46.6340},
    {"nome_local": "Banheiro Unisex 1", "latitude": -23.5810, "longitude": -46.6330},
    {"nome_local": "Banheiro Unisex 2", "latitude": -23.5812, "longitude": -46.6345},
    {"nome_local": "Banheiro Unisex 3", "latitude": -23.5822, "longitude": -46.6325},
    {"nome_local": "Banheiro Unisex 4", "latitude": -23.5833, "longitude": -46.6352},
    {"nome_local": "Banheiro Unisex 5", "latitude": -23.5841, "longitude": -46.6349},
    {"nome_local": "Banheiro Unisex 6", "latitude": -23.5847, "longitude": -46.6328},
    {"nome_local": "Banheiro Unisex 7", "latitude": -23.5853, "longitude": -46.6339},
    {"nome_local": "Área de segurança e primeiros socorros", "latitude": -23.5860, "longitude": -46.6320}
]

# Parâmetros de tempo
inicio = datetime(2025, 11, 15, 15, 0)
fim = datetime(2025, 11, 16, 5, 0)
intervalo = timedelta(minutes=15)

# Lista para armazenar dados
dados = []

# Simulação
timestamp = inicio
while timestamp <= fim:
    for idx, local in enumerate(locais, start=1):
        # Fluxo depende do tipo de local
        if "Palco" in local['nome_local']:
            base_fluxo = np.random.randint(300, 500)
            if 21 <= timestamp.hour <= 23:  # pico
                base_fluxo *= np.random.uniform(3, 4)
        elif "Área de convivência" in local['nome_local']:
            base_fluxo = np.random.randint(200, 400)
        elif "Banheiro" in local['nome_local']:
            base_fluxo = np.random.randint(50, 150)
        else:  # segurança
            base_fluxo = np.random.randint(30, 100)

        dados.append({
            "local_id": idx,
            "nome_local": local['nome_local'],
            "latitude": local['latitude'],
            "longitude": local['longitude'],
            "pessoas_detectadas": int(base_fluxo),
            "hora": timestamp.strftime('%Y-%m-%d %H:%M:%S')
        })
    timestamp += intervalo

# Criar DataFrame
df = pd.DataFrame(dados)

# Exportar CSV
df.to_csv("concentracao_pessoas_evento.csv", index=False)

print("✅ Tabela 'concentracao_pessoas_evento.csv' criada com sucesso!")

Saída da execução
0KB
	Stream
		✅ Tabela 'concentracao_pessoas_evento.csv' criada com sucesso!


```
